```python
data = {'die':44, 'people':23, 'year':11, 'american':11, 'infect':9, 
        'day':9, 'man':8, 'study':8, 'treatment':7, 'start':7}
```

```python
# Create and generate a word cloud image:
wordcloud = WordCloud(width = 100, height = 80, scale = 3.0, 
                      max_font_size = 200, background_color="white")
gen = wordcloud.generate_from_frequencies(data)

# Display the generated image:
plt.figure() 
plt.axis("off")
plt.imshow(gen, interpolation='bilinear')

# Save the image in the img folder
#wordcloud.to_file("img/first_review.png")
```
